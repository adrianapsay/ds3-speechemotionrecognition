```python
PATH = '../data/Crema/1001_DFA_ANG_XX.wav'
waveform, sample_rate = torchaudio.load(PATH)
```

```python
def plot_waveform(waveform, sample_rate):
    waveform = waveform.numpy()

    num_channels, num_frames = waveform.shape
    time_axis = torch.arange(0, num_frames) / sample_rate

    figure, axes = plt.subplots(num_channels, 1)
    if num_channels == 1:
        axes = [axes]
    for c in range(num_channels):
        axes[c].plot(time_axis, waveform[c], linewidth=1)
        axes[c].grid(True)
        if num_channels > 1:
            axes[c].set_ylabel(f"Channel {c+1}")
    figure.suptitle("waveform")
```

```python
plot_waveform(waveform, sample_rate) # plot the waveform of a Crema audio
```

```python
wv_arr = waveform.numpy()[0]
wv_arr.shape
```

```python
fft = np.fft.fft(wv_arr, norm='ortho')
fs_shifted = np.linspace(-len(fft) // 2, len(fft) // 2, len(fft)) * sample_rate / len(fft)

plt.plot(fs_shifted, np.abs(np.fft.fftshift(fft)))
plt.show()
```

```python
peak_thres = fft.copy()
peak_thres[peak_thres < 0.03] = 0

plt.plot(fs_shifted, np.abs(np.fft.fftshift(peak_thres))) # plots audio signal with applied fft
```

Potential Feature Extraction (for emotion recognition)

1) Zero Crossing Rate (how frequently the signal changes its sign/crosses the zero amplitude axis)
* Can potentially help show emotion since higher zero crossing rates reflect high-frequency sounds and lower rates reflect low-frequency sounds (emotions such as anger (ANG), happy (HAP) could be high-frequency and fear (FEA), sadness (SAD), disgust (DIS) can be low frequency)

```python
p1, sr = librosa.load('../data/Ravdess/audio_speech_actors_01-24/Actor_01/03-01-05-01-01-01-01.wav')
FRAME_LENGTH = 1023
HOP_LENGTH = 512
```

```python
zcr_p1 = librosa.feature.zero_crossing_rate(p1, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
```

```python
frames = range(len(zcr_p1))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 10))

plt.plot(t, zcr_p1, color="y")
plt.ylim(0, 1)
plt.xlabel('Time (s)')
plt.ylabel('Zero Crossing Rate')
plt.title('Zero Crossing Rate of the Audio Signal')
plt.show()
```

Depicts noise floor

At 2 sec, close to 0 zero-crossing rate, indicating constant polarity !

Next, find RMSE (to check the effectiveness of noise reduction based on zero crossing rate)

```python
rms_p1 = librosa.feature.rms(y=p1, frame_length = FRAME_LENGTH, hop_length = HOP_LENGTH)[0]
frames = range(len(rms_p1))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)
```

```python
plt.figure(figsize=(15, 17))

ax = plt.subplot(3, 1, 1)
librosa.display.waveshow(p1, alpha=0.5)
plt.plot(t, rms_p1, color="r")
plt.ylim((-1, 1))
plt.title("P1")
```

Next, find pitch over time

```python
pitches, magnitudes = librosa.core.piptrack(y=p1, sr=sample_rate)
times = librosa.times_like(pitches, sr=sample_rate)

plt.plot(times, pitches.T, label='Pitch (Hz)')

plt.title('Pitch of the Audio Signal Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Pitch (Hz)')
plt.legend()
plt.show() #median pitch is 0?
```

```python
plt.plot(times, np.mean(pitches, axis=0), label='Mean Pitch (Hz)', color='b')
plt.title('Pitch of the Audio Signal Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Pitch (Hz)')
plt.legend()
plt.show()
```

Looks like the zero crossing rate (rapid changes in amplitude vs freuqency count)

Making a spectogram to see time + frequency + pitch

```python
pitches, magnitudes = librosa.piptrack(y=p1, sr=sample_rate)
plt.subplot(212)
plt.imshow(pitches[:100, :], aspect="auto", interpolation="nearest", origin="lower")
plt.title('pitches')
```

Frequency up, pitch up

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512

def amplitude_envelope(signal, frame_size, hop_length):
    #Calculate the amplitude envelope of a signal with a given frame size and hop length.
    amplitude_envelope = []
    
    # calculate amplitude envelope for each frame
    for i in range(0, len(signal), hop_length): 
        amplitude_envelope_current_frame = max(signal[i:i+frame_size]) 
        amplitude_envelope.append(amplitude_envelope_current_frame)
    
    return np.array(amplitude_envelope) 


ae_p1 = amplitude_envelope(p1, FRAME_SIZE, HOP_LENGTH)
len(ae_p1)


frames = range(len(ae_p1))
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 17))

ax = plt.subplot(3, 1, 1)
librosa.display.waveshow(p1, alpha=0.5)
plt.plot(t, ae_p1, color="r")
plt.ylim((-1, 1))
plt.title("p1")


plt.show()
```

```python
# Compute pitch and magnitudes
pitches, magnitudes = librosa.core.piptrack(y=p1, sr=sample_rate)

# Extract the pitch values for the first frame
pitch_values = pitches[0, :]

# Plot pitch values against time
times = librosa.times_like(pitch_values, sr=sample_rate)
plt.plot(times, pitch_values, label='Pitch (Hz)', color='r')
plt.title('Pitch of the Audio Signal')
plt.xlabel('Time (s)')
plt.ylabel('Pitch (Hz)')
plt.legend()
plt.show()
```

```python
# get wav from Crema
wav_files = os.listdir('../data/Crema')
filename_parts = pd.Series(wav_files).str.split("_")

wav_info = pd.DataFrame(filename_parts.tolist(), columns=['id', 'val', 'emotion', 'version']).drop(columns="val")
wav_info['file name'] = wav_files

wav_info
```

```python
# get wav from Crema

# Define the folder containing the .wav files
folder_path = '../data/Crema'
wav_val = pd.DataFrame({'id': [], 'Sample rate':[],'Waveform shape':[],'Waveform':[]})

# Iterate over all files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".wav"):
        file_path = os.path.join(folder_path, file_name)
        # Load the .wav file using torchaudio.load()
        waveform, sample_rate = torchaudio.load(file_path)
        
        filename = os.path.basename(file_path)
        filename_parts = filename.split("_")
        
        crema = pd.DataFrame({'id': [filename_parts[0]],'Sample rate': sample_rate, 'Waveform shape': [waveform.shape[0]],'Waveform':[waveform.tolist()]})
        
        wav_val = pd.concat([wav_val,crema])
                              
```
